# Create SQL tools in Unity Catalog

```python
%sql
CREATE OR REPLACE FUNCTION cindy_demo_catalog.movies.lookup_movie_economics(movie_name STRING COMMENT 'name of the movie to look up')
RETURNS TABLE(tmdbId STRING, title STRING, budget DOUBLE, revenue DOUBLE)
COMMENT 'This tool takes movie name(s) as an input and returns movie box office information including budget and revenue.'
RETURN 
  SELECT tmdbId, title, budget, revenue
    FROM cindy_demo_catalog.movies.movielens_small
    WHERE lower(title) LIKE concat('%', lower(movie_name), '%')
;
```

```python
%sql
CREATE OR REPLACE FUNCTION cindy_demo_catalog.movies.lookup_movie_rating(movie_name STRING COMMENT 'name of the movie to look up')
RETURNS TABLE(tmdbId BIGINT, title STRING, rating STRING)
COMMENT 'This tool takes movie name(s) as an input and returns movie box office information including budget and revenue.'
RETURN 
  SELECT
      tmdbId,
      title,
      certification AS rating
    FROM cindy_demo_catalog.movies.movielens_small
    WHERE lower(title) LIKE concat('%', lower(movie_name), '%')
;
```
